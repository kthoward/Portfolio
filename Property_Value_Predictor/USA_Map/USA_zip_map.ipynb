```python
import folium
import pandas as pd
import matplotlib.colors as colors
import matplotlib.cm as cmx
import matplotlib.cm as mplcm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Creating the zip code distribution maps and prediction maps

### Loading in the data

```python
# Loading in the zipcodes from modeling
test_set = pd.read_csv('test_set_final.csv')
```

```python
# Loading in the gain versus loss probability metric
tvm_metric = pd.read_csv('tvm_metric.csv')
tvm_metric = tvm_metric.set_index('zips')
```

```python
# Loading in the latitude and longitude
lat_lon = pd.read_csv('lat_lon.csv')
```

```python
# Merging the data frames
data_to_graph = pd.merge(test_set, lat_lon, left_index=True, right_index=True, how='inner')
```

```python
data_to_graph = data_to_graph.set_index('Unnamed: 0')
```

```python
data_to_graph = pd.merge(data_to_graph, tvm_metric, left_index=True, right_index=True, how='inner')
```

```python
data_to_graph_2 = data_to_graph.sort_values(['top_vs_mean_low'], ascending=False)
```

```python
pred = len(data_to_graph_2) * .1
pred = round(pred)
data_to_graph_2 = data_to_graph_2.iloc[:pred, :]
```

```python
data_to_graph_2.head()
```

```python
target = data_to_graph['target']
```

```python
# Normalizing the predictive metric
ratio = data_to_graph['top_vs_mean_low'] + .9
```

#### Creating a US graph based on zip codes, prediction confidence, and outcome

```python
folium_map_2 = folium.Map(location=[40.4405556, -79.9961111],
                        zoom_start=12,
                        tiles='cartodbpositron')
for index, row in data_to_graph.iterrows():
    target = row['target']
    print(target)
    ratio = row['top_vs_mean_low'] + 2
    radius = ratio ** 2.75
    c = ((target - .9) * 1050)
    c = int(c)
    rgb_values = cm(c)
    print(rgb_values)
    print(c)
    rgb_values = np.array(rgb_values[:3])
    rgb_values = rgb_values * 255
    rgb_values = rgb_values.astype(int)
    print(rgb_values)
    color = "#{:02x}{:02x}{:02x}".format(rgb_values[0],rgb_values[1],rgb_values[2])
    print(color)
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius=radius,
                        color=color,
                        fill=True).add_to(folium_map_2)
```
